```python
experiencia = [1, 3, 5, 7, 9]
salario_data_scientists = [3434, 6436, 2343, 4656, 8999]
salario_ing_software = [3344, 63446, 2343, 46456, 84599]
```

```python
# Grafica la relación entre experiencia y salario para Data Scientists
plt.plot(experiencia, salario_data_scientists)

# Grafica la relación entre experiencia y salario para Ingenieros de Software
plt.plot(experiencia, salario_ing_software)

# Añade un título al gráfico
plt.title('Salarios')

# Añade una etiqueta al eje x
plt.xlabel('Experiencia')

# Añade una etiqueta al eje y
plt.ylabel('Ganancias')

# Muestra el gráfico
plt.show()
```

```python
# Grafica la relación entre experiencia y salario para Data Scientists
plt.plot(experiencia, salario_data_scientists, color = "r",label = "Data scientist")
plt.plot(experiencia, salario_data_scientists, color = "g",label = "Ing Software")

# Grafica la relación entre experiencia y salario para Ingenieros de Software
plt.plot(experiencia, salario_ing_software)

# Añade un título al gráfico
plt.title('Salarios')

# Añade una etiqueta al eje x
plt.xlabel('Experiencia')

# Añade una etiqueta al eje y
plt.ylabel('Ganancias')

# Añade una leyenda para identificar las dos líneas en el gráfico
plt.legend(['Data scientist', 'Ing Software'])

# Muestra el gráfico
plt.show()
```

```python
# Grafica la relación entre experiencia y salario para Data Scientists con opciones de estilo
plt.plot(experiencia, salario_data_scientists, color='#0f0f0f', label='Data scientist', linestyle='--', marker='o')

# Grafica la relación entre experiencia y salario para Ingenieros de Software con opciones de estilo
plt.plot(experiencia, salario_ing_software, color='g', label='Ing Software', linestyle=':', marker='*')

# Añade un título al gráfico
plt.title('Salarios')

# Añade una etiqueta al eje x
plt.xlabel('Experiencia')

# Añade una etiqueta al eje y
plt.ylabel('Ganancias')

# Añade una leyenda para identificar las dos líneas en el gráfico
plt.legend()

# Añade una cuadrícula al gráfico
plt.grid(True)

# Muestra el gráfico
plt.show()
```

```python
# Define los datos de las categorías y sus valores
x = ['A', 'B', 'C', 'D']
y = [3, 8, 1, 10]

# Crea un gráfico de barras usando los datos de x e y
plt.bar(x, y)

# Añade un título al gráfico
plt.title('Salarios')

# Añade una etiqueta al eje x
plt.xlabel('Experiencia')

# Añade una etiqueta al eje y
plt.ylabel('Ganancias')

# Muestra el gráfico
plt.show()
```

```python

datos=np.random.rand(1000)

plt.hist(datos, bins=79, color='navy')
plt.title('Histograma')
plt.xlabel( 'Valor')
plt.ylabel('Frecuencia')
```

```python
# Genera tres conjuntos de datos usando una distribución normal con diferentes desviaciones estándar
# np.random.normal(0, std, 100) genera 100 valores con una media de 0 y una desviación estándar de 'std'
# for std in range(1, 4) itera sobre los valores 1, 2 y 3 para 'std'
datos = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Crea un diagrama de caja usando los conjuntos de datos generados
# plt.boxplot(datos, vert=True) crea un diagrama de caja vertical
plt.boxplot(datos, vert=True)

# Muestra el gráfico
# plt.show() muestra el gráfico en pantalla
plt.show()
```

```python


# Datos para la gráfica de líneas
experiencia = [1, 3, 5, 7, 9]
salario_data_scientists = [3434, 6436, 2343, 4656, 8999]
salario_ing_software = [3344, 63446, 2343, 46456, 84599]

# Datos para la gráfica de barras
categorias = ['A', 'B', 'C', 'D']
valores = [3, 8, 1, 10]

# Datos para la gráfica de dispersión
x3 = np.random.rand(100)
y3 = np.random.rand(100)

# Datos para el diagrama de caja
datos = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Crear una figura con una cuadrícula de 2x2 subgráficas
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Gráfica de líneas
axs[0, 0].plot(experiencia, salario_data_scientists, color='#0f0f0f', label='Data scientist', linestyle='--', marker='o', linewidth=2)
axs[0, 0].plot(experiencia, salario_ing_software, color='g', label='Ing Software', linestyle=':', marker='*', linewidth=2)
axs[0, 0].set_title("Gráfica de líneas")
axs[0, 0].set_xlabel('Experiencia')
axs[0, 0].set_ylabel('Ganancias')
axs[0, 0].legend()
axs[0, 0].grid(color='r', linestyle='--', linewidth=2)

# Gráfica de barras
axs[0, 1].bar(categorias, valores)
axs[0, 1].set_title("Gráfico de barras")
axs[0, 1].set_xlabel('Categorías')
axs[0, 1].set_ylabel('Valores')

# Gráfica de dispersión
axs[1, 0].scatter(x3, y3, color='r', alpha=0.5)
axs[1, 0].set_title('Gráfico de dispersión')
axs[1, 0].set_xlabel('Eje X')
axs[1, 0].set_ylabel('Eje Y')

# Diagrama de caja
axs[1, 1].boxplot(datos, vert=True)
axs[1, 1].set_title('Diagrama de caja')
axs[1, 1].set_xlabel('Conjuntos de datos')
axs[1, 1].set_ylabel('Valores')
```
