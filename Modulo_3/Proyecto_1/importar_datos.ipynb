# Extrae numeros de una columna

```python
import pandas as pd
import re
import numpy as np
data = {'mesclados' : ['fdgdg3', 'xvrfv4', 'fdawer43','4333fcx']}
df = pd.DataFrame(data)
```

```python
data_num ='fdgdg3'

numero = re.search(r'\d+',data_num)
int(numero.group())
```

```python
def extrac_numbers(s):
    numero = re.search(r'\d+',s)
    return int(numero.group()) if numero else np.nan
```

```python
df['mesclados'] = df['mesclados'].apply(extrac_numbers)
```

# Deteccion de valores atipicos

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
Q1 = df_clean['Age'].quantile(0.25)
Q3 = df_clean['Age'].quantile(0.75)

IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR

limite_inferior

limite_superior = Q3 + 1.5 * IQR
limite_superior
```

# Notas : simulacion de datos atipicos

```python
df_test = df_clean.copy()

indices = np.random.choice(df_test.index, 2, replace=False)

df_test.loc[indices[0], 'Age'] += 3 * IQR
df_test.loc[indices[1], 'Age'] += 3 * IQR
```

```python
df_test.boxplot(column='Age')
```
